```python
import matplotlib.pyplot as plt

# Example data: list of points with timestamps
timestamps = ["12:00", "12:05", "12:10", "12:15", "12:20"]
x_points = [0, 0.5, 1.0, 1.5, 2.0]
y_points = [0.1, 1.0, 0.5, 1.5, 1.0]

# Create a figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# --- Plot 1: Points with timestamps ---
axes[0].scatter(x_points, y_points, c='blue')
for i, t in enumerate(timestamps):
    axes[0].text(x_points[i]+0.05, y_points[i]+0.05, str(t), fontsize=9)
axes[0].set_title("Point Storage")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].grid(True)
axes[0].set_ylim(0, max(y_points)+0.2)
axes[0].set_xlim(-0.1, max(x_points)+0.2)
# --- Plot 2: Trajectory with timestamps ---
axes[1].plot(x_points, y_points, marker='o', color='blue')
for i, t in enumerate(timestamps):
    axes[1].text(x_points[i]+0.05, y_points[i]+0.05, str(t), fontsize=9)
axes[1].set_title("Trajectory Storage")
axes[1].set_xlabel("X")

axes[1].set_ylabel("Y")
axes[1].grid(True)
axes[1].set_ylim(0, max(y_points)+0.2)
axes[1].set_xlim(-0.1, max(x_points)+0.2)

plt.tight_layout()
plt.savefig("trajectory_representation_options.pdf", bbox_inches='tight')
plt.show()
```
